What is the average selling price of cars for each dealer, and how does it compare across different dealers?

```python
average_prices = df.groupby("Dealer_Name")["Price ($)"].mean()

print("Average selling price for each dealer")
print(round(average_prices,2))

average_prices.plot(kind = 'bar',color="skyblue")
plt.xlabel("Dealer Name")
plt.ylabel("Average Selling Price ($)")
plt.show()
```

Which car brand (Company) has the highest variation in prices, and what does this tell us about the pricing
trends?

-->INSIGHTS =The car company with the highest variation in prices indicates that there is a wide range of prices for cars from that company. This could mean that the company offers a diverse range of models with different features and price points, catering to various market segments.

This variation shows that Brands(Cadillac,Lincoln,Plymoyth,Saab) is catering to a broad audience with different needs and budgets.

```python
#by calculating standard deviation,The brand with the highest standard deviation
#has the highest variation in prices.
price_variation = df.groupby("Company")["Price ($)"].std()
```

```python
# A line plot alone may not provide a clear picture of the variation.
plt.figure(figsize=(12,8))
sns.lineplot(x ="Company",y="Price ($)",data = df)
plt.xticks(rotation = 45)
plt.show()
```

```python
## A Bar plot to provide a clear picture of the variation.
plt.figure(figsize=(16,8))
price_variation.plot(kind = "bar",color = "darkgreen")
plt.xlabel("Company")
plt.ylabel("Price_variation")
plt.xticks(rotation = 45)
plt.show()
```

What is the distribution of car prices for each transmission type, and how do the interquartile ranges compare?

-->From the comparison of IQR, we can conclude that the average prices of both manual and automatic cars are approximately the same. There is a slight increase in the prices of automatic cars compared to manual cars. Additionally, for 75% of automatic cars, the prices are similar, but in the highest price segment, manual cars are more expensive than automatic cars.

4.What is the distribution of car prices across different regions?

```python
plt.figure(figsize=(12,8))
sns.histplot(x="Price ($)",hue="Dealer_Region",kde = True,data=df,bins=30)
plt.title("distribution of car prices across different regions")
plt.xlabel("Price ($)")
plt.ylabel("frequency")
plt.show()
```

6. How does the average selling price of cars vary by customer gender and annual income?

```python
avg_sell_price = df.groupby(["Gender","Annual Income"])["Price ($)"].mean().reset_index()

print("average selling price of cars vary by customer gender and annual income")
print(avg_sell_price)
plt.figure(figsize=(12,8))
sns.barplot(x="Annual Income",y="Price ($)",hue="Gender",data=avg_sell_price)
plt.show()
```

7.What is the distribution of car prices by region, and how does the number of cars sold vary by region?

```python
plt.figure(figsize=(12,8))
plt.subplot(2,1,1)
sns.histplot(x="Dealer_Region",hue ="Dealer_Region",data=df)
plt.title('distribution of car prices by region')
plt.tight_layout()


plt.figure(figsize=(12,8))
cars_sold_by_region = df['Dealer_Region'].value_counts().reset_index()
cars_sold_by_region.columns = ['Dealer_Region', 'Number of Cars Sold']

plt.subplot(2, 1, 2)
sns.barplot(x='Dealer_Region', y='Number of Cars Sold', data=cars_sold_by_region, palette="viridis")
plt.title('Number of Cars Sold by Region')
plt.xlabel('Dealer Region')
plt.ylabel('Number of Cars Sold')
plt.tight_layout()#Ensures that the subplots do not overlap and are displayed clearly.
plt.show()
```

8.How does the average car price differ between cars with different engine sizes

```python
differ_price = df.groupby("Engine")["Price ($)"].mean().reset_index()
sns.lineplot(x='Engine',y="Price ($)",data=differ_price)
```

How does car price vary with engine size across different car colors, and which colors have the highest price
variation

```python
avg_price = df.groupby(["Engine","Color"])["Price ($)"].mean().reset_index()
price_variation = df.groupby("Color")["Price ($)"].std().reset_index()

plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
sns.barplot(x="Engine",y="Price ($)",hue="Color",data=avg_price)
plt.show()

plt.figure(figsize=(12,8))
plt.subplot(1,2,2)
sns.barplot(x="Color",y="Price ($)",data=price_variation)
plt.show()
```
