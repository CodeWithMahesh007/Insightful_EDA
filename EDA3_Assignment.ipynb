```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("amazon.csv")
df
```

1. What is the average rating for each product category?

```python
df.groupby("category")["rating"].mean()
#here rating is object data type but it should float,num we convert it into float data type.
```

```python
df["rating"]=df.rating.str.replace('|',"0")
```

```python
df.rating.unique()
```

```python
df["rating"]=df.rating.astype("float")
```

```python
df.rating.dtype
```

2. What are the top rating_count products by category?

```python
df.groupby("category")["rating"].count().sort_values(ascending = False).reset_index()
```

3. What is the distribution of discounted prices vs. actual prices?

```python
sns.pairplot(df[['discounted_price', 'actual_price']], kind='scatter', diag_kind='kde')
```

```python
df["actual_price"]=df.actual_price.str.replace("₹","").str.replace(",","")
```

```python
#convert to numeric
df["actual_price"] = pd.to_numeric(df["actual_price"])
```

```python
#convert to integer
df["actual_price"] = df["actual_price"].astype(int)
```

```python
print(df["actual_price"].dtype)
```

```python
#same with discounted price.
df["discounted_price"]=df.discounted_price.str.replace("₹","").str.replace(",","")
```

```python
df["discounted_price"] = pd.to_numeric(df["discounted_price"])
df["discounted_price"] = df["discounted_price"].astype(int)
```

```python
print(df["discounted_price"].dtype)
```

4. What are the most popular product names?

```python
popular_products=df["product_name"].value_counts().reset_index()
popular_products.columns = ["product_name","rating_count"]
print(popular_products)
```

5.What are the most popular product reviews?

```python
# Count the number of occurrences for each review
popular_reviews = df['review_content'].value_counts().reset_index()

# Rename the columns for clarity
popular_reviews.columns = ['review_content', 'review_count']

# Display the most popular product reviews
print(popular_reviews)
```

6. What is the correlation between discounted_price and rating?

```python
desired_columns = ["discounted_price","rating"]
correlation_matrix =df[desired_columns].corr()
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',linewidth=0.5)
```
